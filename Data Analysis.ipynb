# WHAT DO WE DO TO MISSING VALUES:  
`1.DROP MISSING VALUES` - This is the fastest and easiest step to handle missing values. However, it is not generally advised. This method reduces the quality of our model as it reduces sample size because it works by deleting all other observations where any of the variable is missing.   


```python
#DROP MISSING VALUES BY :-
#data.dropna()
```

`2.FILL MISSING VALUES WITH TEST STATISTIC`-  
This is a process whereby missing values are replaced with a test statistic like mean, median or mode of the particular feature the missing value belongs to.


```python
# for col in data.columns:
#     if data[col].isnull().any():
#         median = data[col].median()
#         data[col] = data[col].fillna(median)
        
#         data[col]=data[col].fullna(median)
```

```python
for col in data.columns.values:
    if data[col].isnull().sum() ==0:
        continue
        if col == 'Climate':
            guess_values = data.groupby('Region')['Climate'].apply(lambda x: x.mode().max())
            print(1)
        else:
            guess_values = data.groupby('Region')[col].median()
        for region in data['Region'].unique():
            data[col].loc[(data[col].isnull())&(data['Region']==region)] = guess_values[region]
```
