# Cleaning the data:
The data has "," seperator in the numerical columns which has to be replaced by "."  
A fuction to do that would be as follows:

```python
for column in data.columns:
    data[column]= data[column].astype(str)
    if column != 'Country' and column!= 'Region':
        for row in range(len(data[column])):
            if "," in data[column][row]:
                data[column][row]=".".join(data[column][row].split(","))
                print(data[column][row])
```

```python
data
```

```python
data.dtypes
```

```python
# converting data type to float
for column in data.columns:
    if column != "Country" and column != "Region":
        data[column]=pd.to_numeric(data[column] , errors = 'coerce')
```

```python
#DROP MISSING VALUES BY :-
data.dropna(inplace= True)
```

## Modelling 

Importing Label Encoder from the SKLEARN lib

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
from sklearn.model_selection import train_test_split
```

```python
data['Regional'] = le.fit_transform(data['Region'])
data['Climate'] = le.fit_transform(data['Climate'])
data.dropna()
```

```python
train, test = train_test_split(data, test_size=0.3, shuffle=True)
```

```python

training_features = ['Population', 'Area (sq. mi.)',
       'Pop. Density (per sq. mi.)', 'Coastline (coast/area ratio)',
       'Net migration', 'Infant mortality (per 1000 births)',
       'Literacy (%)', 'Phones (per 1000)',
       'Arable (%)', 'Crops (%)', 'Other (%)', 'Birthrate',
       'Deathrate', 'Agriculture', 'Industry', 'Service','Service']
target = 'GDP ($ per capita)'
```

```python
train_X=train[training_features]
train_y=train[target]
test_X=test[training_features]
train_Y=test[target]
```

Using linear regression Model

```python
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
```
